```python
from pathlib import Path
from sqlalchemy.types import Date, Float, Integer, Text
import pandas as pd

# Correct path to your CSV file inside the project
csv_path = Path("Healthcare_Data_Project/data_files/medical_data.csv")

# Load the dataset into a pandas DataFrame
df = pd.read_csv(csv_path)

# Optional: preview the first few rows
df.head()
```

```python
## Load Data into PostgreSQL (Raw Schema)

from sqlalchemy import create_engine

# Replace with your PostgreSQL credentials
engine = create_engine("postgresql+psycopg2://dquser:dqpass@localhost:5432/healthcare")

df.to_sql(
    "patients_raw",
    engine,
    schema="raw",
    if_exists="replace",   # replaces table if it already exists
    index=False
)

with engine.connect() as conn:
    result = conn.execute("SELECT table_schema, table_name FROM information_schema.tables WHERE table_schema='raw';")
    print(result.fetchall())
```
